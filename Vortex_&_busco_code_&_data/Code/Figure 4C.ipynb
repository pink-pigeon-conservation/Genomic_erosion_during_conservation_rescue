```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

```python
dfh = pd.read_csv('Github_code_data/Data/Fig_4c.csv', index_col = 0)
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
heat = sns.heatmap(dfh, cmap="RdYlGn_r")
heat.invert_yaxis()
heat.figure.axes[-1].set_ylabel('Probability of extinction(PE)', rotation = 270, labelpad = 25, size = 14) # seaborn doesn't give access to objects but it the colour bar was the last axes you created so you can modify it that way
heat.set_xlabel('#LE', labelpad = 15, fontsize = 14)
heat.set_ylabel('Inbreeding(F)', labelpad = 15, fontsize = 14)
heat.grid(b = True, which = 'both')

plt.savefig("Heatmap_PE_100_Yr", format='pdf', dpi=1000)

```
